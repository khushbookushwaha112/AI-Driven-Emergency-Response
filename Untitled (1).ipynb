```python
# AI ko sikhane ke liye keywords
crime_keywords = {
    "Serious Road Accident": ["accident", "hit", "crash", "road", "injury", "khoon", "blood"],
    "Kidnapping": ["kidnap", "abduct", "missing", "gayab", "utha liya", "child"],
    "Armed Robbery": ["gun", "robbery", "pistol", "loot", "weapon", "chaku"],
    "Chain/Mobile Snatching": ["snatch", "cheen", "snatching", "bhag gaya", "bike"]
}

print("AI Vocabulary Ready!")
```

```python
def improved_ai_parser(fir_text):
    text = fir_text.lower()
    detected_crime = "Unknown"

    # Keyword matching logic
    for crime, keywords in crime_keywords.items():
        if any(word in text for word in keywords):
            detected_crime = crime
            break

    # Location detection (Simple)
    locations = ["rohini", "saket", "dwarka", "noida", "gurgaon", "lajpat nagar"]
    detected_loc = "Unknown Location"
    for loc in locations:
        if loc in text:
            detected_loc = loc.capitalize()
            break

    return detected_crime, detected_loc

print("Improved Parser Ready!")
```

```python
def final_emergency_trigger(fir_input):
    # 1. AI Analysis
    crime, loc = improved_ai_parser(fir_input)

    if crime == "Unknown":
        print(" AI ERROR: Could not identify crime. Alerting Human Supervisor.")
        return

    # 2. Get Rules from Step 2 (df_crime_matrix)
    # Hum 'crime' variable ko direct match karenge
    logic = df_crime_matrix[df_crime_matrix['Crime_Type'] == crime].iloc[0]

    print(f" --- AI ANALYSIS REPORT ---")
    print(f"REPORTED CASE: {crime}")
    print(f"LOCATION     : {loc}")
    print(f"PRIORITY     : {logic['Priority']}")
    print("-" * 35)

    # 3. Smart Medical Check (Step 3 Integration)
    # Agar accident, murder ya violence hai toh hospital ko alert do
    if any(x in crime.lower() for x in ["accident", "murder", "violence"]):
        print("\n [MEDICAL ALERT]: High Priority!")
        # Humne Step 3 mein jo hospitals banaye the, wahan se pehla hospital uthate hain
        hospital = df_infra_complete[df_infra_complete['Type'] == 'Hospital'].iloc[0]
        print(f"SENDING EMERGENCY SIGNAL TO: {hospital['Name']}")
        print(f"HOSPITAL ACTION: ICU & Trauma team on standby.")

    # 4. Police Dispatch (Step 1 Integration)
    print(f"\n [POLICE DISPATCH]: {logic['Response_Units']} sent to {loc}.")
    print(f"PROTOCOL: {logic['Action']}")

print("Full Integration Engine Ready!")
```

```python
import pandas as pd
import numpy as np
import joblib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier

# 1. Keywords Dictionary (System ko sikhane ke liye)
scenarios = {
    "Kidnapping": ["bacha missing", "kidnap", "child abducted", "gayab hai", "car mein le gaye", "van mein utha liya", "minor missing", "abduction reported", "kidnapped from park"],
    "Serious Road Accident": ["road accident", "car crash", "truck accident", "accident ho gaya", "khoon nikal raha hai", "ambulance needed", "major collision", "hit and run", "highway accident"],
    "Armed Robbery": ["gunpoint loot", "armed robbery", "weapon dikhakar", "pistol robbery", "dukan mein loot", "stole cash at gunpoint", "armed men entered", "shastra loot"],
    "Cyber Crime": ["online fraud", "account hack", "money stolen online", "otp scam", "bank fraud", "phishing link", "account se paise nikal gaye", "digital scam"],
    "Murder": ["dead body found", "murder reported", "kisine maar diya", "dead person", "homicide", "goli maar di", "body in the park", "murder with knife"]
}

# 2. Automatic Data Generator (Synthetic Data)
aug_text, aug_label = [], []
for crime, keywords in scenarios.items():
    for _ in range(500):  # Har category ke 500 samples
        # Randomly pick 2-3 words to create a pattern
        selected = np.random.choice(keywords, size=2, replace=False)
        sentence = f"Emergency! {selected[0]} happened. {selected[1]} reported near my location."
        aug_text.append(sentence)
        aug_label.append(crime)

df_pro = pd.DataFrame({'text': aug_text, 'label': aug_label})

# 3. Best Vectorization (N-grams is the Secret Sauce)
# Isse model "car mein le gaye" ko ek poora phrase samajhta hai
vectorizer = TfidfVectorizer(ngram_range=(1, 3), max_features=10000)
X = vectorizer.fit_transform(df_pro['text'])
y = df_pro['label']

# 4. Powerful Model
ml_model = RandomForestClassifier(n_estimators=300, max_depth=None, random_state=42)
ml_model.fit(X, y)

# 5. Save Files
joblib.dump(ml_model, 'crime_model.pkl')
joblib.dump(vectorizer, 'vectorizer.pkl')

print(" SUCCESS: Model is now trained on 2,500+ Scenarios!")
```

```python
import joblib

# 1. Saved files ko load karein
loaded_model = joblib.load('crime_model.pkl')
loaded_vect = joblib.load('vectorizer.pkl')

def test_ml_model():
    print(" ML MODEL TESTING SUITE")
    print("-------------------------")

    # User se input lein
    test_input = input(" Enter a test sentence: ")

    if test_input.strip() == "":
        print(" Empty input!")
        return

    # 2. Text ko vectorizer se transform karein
    text_vector = loaded_vect.transform([test_input])

    # 3. Predict karein
    prediction = loaded_model.predict(text_vector)[0]

    # 4. Confidence Score (Optional: Model kitna sure hai)
    probability = loaded_model.predict_proba(text_vector).max() * 100

    print(f"\n AI Prediction: **{prediction}**")
    print(f" Confidence: {probability:.2f}%")
    print("-" * 25)

# Testing shuru karein
test_ml_model()
```
